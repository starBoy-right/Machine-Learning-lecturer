```python
data_cuaca['Temperatur Udara (°C)'].to_numpy(), data_cuaca['Kecepatan Angin (km/jam)'].to_numpy()
```

```python
test_data = (16, 3)
```

```python
(data_cuaca['Temperatur Udara (°C)'].to_numpy() - test_data[0]) + (data_cuaca['Kecepatan Angin (km/jam)'].to_numpy() - test_data[1])
```

```python
(data_cuaca['Temperatur Udara (°C)'].to_numpy() - test_data[0])**2 + (data_cuaca['Kecepatan Angin (km/jam)'].to_numpy() - test_data[1])**2
```

```python
def KNN_Algorthm(data, test_data, k=3):
    # mengubah kedalam numpy agar perhitungannya cepat 
    feature_1 = data['Temperatur Udara (°C)'].to_numpy()
    feature_2 = data['Kecepatan Angin (km/jam)'].to_numpy()

    jarak = (data['Temperatur Udara (°C)'].to_numpy() - test_data[0])**2 + (data['Kecepatan Angin (km/jam)'].to_numpy() - test_data[1])**2
    
    # kita urutkan nilai k
    target = data['Klasifikasi atau Persepsi Marry']
    
    # mencari index data berdasarkan index jarak yang telah disorting
    result = pd.DataFrame({'jarak' : jarak, 'target': target})

    return result.sort_values(by='jarak')[:k]
```

```python
result = KNN_Algorthm(data_cuaca, test_data)
result
```
