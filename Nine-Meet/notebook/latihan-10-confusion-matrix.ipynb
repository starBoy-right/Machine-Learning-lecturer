```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({
    "NIM": ["TI001","TI002","TI003","TI004","TI005","TI006","TI007","TI008","TI009","TI010"],
    "Hasil Sebenarnya": [
        "Lulus","Lulus","Lulus","Lulus","Lulus",
        "Tidak Lulus","Tidak Lulus","Tidak Lulus","Tidak Lulus","Tidak Lulus"
    ],
    "Hasil Prediksi": [
        "Lulus","Lulus","Lulus","Tidak Lulus","Tidak Lulus",
        "Lulus","Tidak Lulus","Tidak Lulus","Tidak Lulus","Tidak Lulus"
    ]
})
```

```python
df
```

```python
# membuat confusion matrix
from sklearn.metrics import confusion_matrix, classification_report
```

1. Confusion Matrix

```python
confusion_matrix(df['Hasil Sebenarnya'], df['Hasil Prediksi'], labels=['Lulus', 'Tidak Lulus'])
```

2. Classification Report

```python
print(classification_report(df['Hasil Sebenarnya'], df['Hasil Prediksi']))
```
