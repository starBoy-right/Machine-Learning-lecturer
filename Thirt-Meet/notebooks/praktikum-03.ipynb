```python
df = df.drop(columns='dteday')
```

```python
# membagi dataset berdasarkan kategorinya, untuk diproses berbeda pada tahap preprocessing

X = df.drop(columns='cnt')
y = df['cnt']
```

### **Membagi dataset train dan test dengan ukuran 20% untuk data testing**

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
bike_model = DecisionTreeRegressor()
bike_model.fit(X_train, y_train)
```

```python
y_pred = bike_model.predict(X_test)
```

```python
# evaluasi model
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error


print(f"nilai r2-score = {r2_score(y_test, y_pred)}")
print(f"nilai mae = {mean_absolute_error(y_test, y_pred)}")
print(f"nilai mse = {mean_squared_error(y_test, y_pred)}")
```

Dari nilai `r2 score` didapatkan informasi bahwa model bagus dan fit dengan data, dilihat dari hasil dimana nilai r2 bernilai 0.99. Hasil ini menunjukkan bahwa SSE (sum of square error) dari model lebih kecil daripada hanya menggunakan SST (total sum of square) atau hany menghitung total kuadrat dari nilai faktual dengan rata-ratanya. sehingga tidak perlu lagi dilakukan feature engineering untuk pemilihan fitur.

Nilai mae yang didapatkan juga tidak terlalu besar atau masih bisa ditolerasi terhadap nilai `cnt` (jumlah peminjaman sepeda baik rental maupun langganan)

jikalau nilai r2 scorenya jelek atau mungkin bernilai sekitar 50 persen, saya kemungkinana akan melakukan feature engineering atau tuning parameter, karena jika dilihat dari data, data bersih dan valid (tidak ada missing value, dll).
